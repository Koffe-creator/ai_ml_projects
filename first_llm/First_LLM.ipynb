```python
from langchain_groq import ChatGroq

try:
    from google.colab import userdata
    GROQ_API_KEY = userdata.get("GROQ_API_KEY")
except ImportError:
    from dotenv import load_dotenv
    import os
    load_dotenv()
    GROQ_API_KEY = os.getenv("GROQ_API_KEY")

llm_groq  = ChatGroq(model_name="llama-3.1-8b-instant", api_key=GROQ_API_KEY, temperature=0.7)
llm_groq2 = ChatGroq(model_name="openai/gpt-oss-120b",   api_key=GROQ_API_KEY, temperature=0.7)
```

```python
ai_mesg = llm_groq2.invoke("When was John Jerry Rawlings of Ghana born")
print(ai_mesg.content)
```
